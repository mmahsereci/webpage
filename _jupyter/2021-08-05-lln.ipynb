# single run

```python
import numpy as np
np.random.seed(42)

# fair coin
p = 0.5

# Bernoulli samples: 1 means heads, 0 means tails
n_samples = int(1e5)
samples = 1 * (np.random.rand(n_samples) < p)

# print the first 10 samples of the first run
samples[:10]
```

```python
S = np.cumsum(samples)
n =  np.arange(1, n_samples + 1)
means = S / n
```

```python
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 15})

n_plot = 100

plt.figure(figsize=(12, 4))
plt.plot(n[:n_plot], means[:n_plot])
plt.hlines(p, 0, n_plot, colors="gray")
# plt.xscale("log")
plt.xlim([1, n_plot])
plt.ylim([0, 1])
plt.xlabel("# of samples")
plt.ylabel("sample mean xn")
plt.show()
```

```python
plt.figure(figsize=(12, 4))
plt.plot(n, means.T, label='sample mean xn')
plt.hlines(p, 0, n_samples, colors="gray", label="p=0.5")
plt.xscale("log")
plt.xlim([1, n_samples])
plt.ylim([0, 1])
plt.xlabel("# of samples")
plt.ylabel("sample mean xn")
# plt.legend()

# plt.savefig(PATH + "00.png", 
#             bbox_inches = 'tight',pad_inches = 0.05)
```

```python
plt.figure(figsize=(12, 4))
plt.plot(n, abs(means - p).T, color="C0")
# plt.hlines(p, 0, n_samples, colors="gray")
plt.xscale("log")
plt.yscale("log")
plt.xlim([1, n_samples])
plt.ylim([0, 1])
plt.xlabel("# of samples")
plt.ylabel("sample mean xn")
plt.show()
```

# multiple runs

```python
n_samples = int(1e5)  # more samples than before
n_runs = 5  # number of runs

# Bernoulli samples: 1 means heads, 0 means tails
samples = 1 * (np.random.rand(n_runs * n_samples) < p)
samples = np.reshape(samples, (n_runs, n_samples))
```

```python
S = np.cumsum(samples, axis=1)
n =  np.arange(1, n_samples + 1)
means = S / n
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 4))
plt.plot(n, means.T, color='C0')
plt.hlines(p, 0, n_samples, colors="gray")
plt.xscale("log")
plt.xlim([1, n_samples])
plt.ylim([0, 1])
plt.xlabel("# of samples")
plt.ylabel("sample mean xn")
plt.show()
```
